## 1.4 Evaluasi Metrik

Beberapa metrik evaluasi yang akan digunakan untuk menilai performa model adalah sebagai berikut:
- **R-Squared (R²)**: Indikasi seberapa baik model mampu menjelaskan variasi data yang ada. Nilai R² yang lebih tinggi menunjukkan model yang lebih baik dalam menjelaskan hubungan antara fitur dan target.

- **MAE (Mean Absolute Error)**: Mengukur rata-rata selisih absolut antara prediksi dan nilai aktual. MAE memberikan gambaran seberapa besar kesalahan prediksi model dalam satuan yang sama dengan target (harga mobil bekas).

- **RMSE (Root Mean Squared Error)**: RMSE memberikan gambaran lebih jelas mengenai dampak kesalahan yang lebih besar dalam prediksi. Metrik ini sensitif terhadap outlier, sehingga sangat berguna dalam evaluasi model yang berpotensi menghasilkan prediksi ekstrem.

- **MAPE (Mean Absolute Percentage Error)**: Mengukur seberapa besar persentase kesalahan dalam prediksi model, memberikan perspektif yang lebih mudah dipahami dalam konteks nilai relatif.

Dari metrik ini, R² akan menjadi metrik utama yang digunakan untuk mengevaluasi keberhasilan model, sementara metrik lainnya (MAE, MSE, RMSE, MAPE) digunakan untuk menggali lebih dalam seberapa besar dan konsisten kesalahan yang dihasilkan oleh model.

**Analisis Korelasi Kolom Price dan Kolom Kategorikal**

```python
def correlation_ratio(categories, values):
    values = np.array(values)
    categories = np.array(categories)
    unique_categories = np.unique(categories)
    category_means = [values[categories == cat].mean() for cat in unique_categories]
    overall_mean = values.mean()
    ss_between = sum(
        len(values[categories == cat]) * (mean - overall_mean) ** 2
        for cat, mean in zip(unique_categories, category_means)
    )
    ss_total = sum((values - overall_mean) ** 2)
    return np.sqrt(ss_between / ss_total)
```

```python
objects = df.select_dtypes(include=['object']).columns

# Menyimpan nilai korelasi
features = []

# Menghitung correlation ratio untuk setiap kolom kategorikal
for col in objects:
    corr = round(correlation_ratio(df[col], df['Price']), 2)
    features.append(corr)

# Membuat DataFrame hasil
df_ratio = pd.DataFrame(
    data=[features], 
    columns=list(objects), 
    index=['Price']
)

# Menampilkan heatmap
plt.figure(figsize=(10, 2))
sns.heatmap(df_ratio, annot=True, cbar=False, cmap='coolwarm')
plt.title('Correlation Ratio of Categorical Features with Price')
plt.show()
```

Berdasarkan hasil korelasi antara kolom katgeorikal dan Price diatas, dapat diketahui bahwa Type dan Make mobil bekas mempengaruhi Harga jual mobil di Syarah.com

#### 4.1.4.2 Penanganan Outlier pada Kolom Mileage

```python
saudi['Mileage'].describe().round(2)
```

Terdapat rentang nilai Mileage yang sangat lebar, yaitu dari 100 hingga 20,000,000. Nilai maksimum yang mencapai 20,000,000 sangat tidak wajar dan jauh lebih tinggi dibandingkan dengan nilai lainnya. Hal ini menunjukkan adanya outlier yang signifikan, yang bisa mempengaruhi analisis atau model yang akan digunakan. Untuk menangani hal tersebut, nilai yang dianggap terlalu jauh dan tidak wajar akan dihapus berdasarkan nilai Upperbound yang telah dihitung berdasarkan rumus berikut:

# 5. Modeling

## 5.1 Train-Test Split

**Train-Test Split** adalah proses membagi dataset menjadi dua bagian utama: train set dan test set. Tujuannya adalah untuk mempersiapkan data agar dapat digunakan untuk melatih dan menguji model machine learning dengan cara yang adil dan efektif.

Dalam proses pembagian dataset, X merepresentasikan fitur-fitur seperti Year, Mileage, Region, dan fitur lainnya sedangkan y adalah variabel target yaitu Price. Dataset dibagi menggunakan fungsi train_test_split dengan rasio 80:20, di mana 80% data digunakan sebagai train set untuk melatih model mempelajari pola dari X untuk memprediksi y, sedangkan 20% data digunakan sebagai test set untuk menguji kemampuan generalisasi model.

```python
X = saudi.drop(columns=['Price'])
y = saudi['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = randomstate)

print(X_train.shape,y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train
```

Pada tahap ini, dilakukan **Feature Selection** menggunakan **metode Variance Inflation Factor (VIF)** untuk mengidentifikasi dan menghapus fitur-fitur yang memiliki multikolinearitas tinggi. Proses ini bertujuan untuk memastikan bahwa hanya fitur yang independen satu sama lain yang digunakan dalam pelatihan model, sehingga meningkatkan interpretabilitas dan performa model.

Untuk meminimalisir data leakage, penghitungan VIF hanya dilakukan pada training set yang telah dipisahkan sebelumnya. Dengan cara ini, informasi dari test set tidak memengaruhi proses pemilihan fitur. Setelah fitur dengan VIF tinggi dihapus berdasarkan analisis pada training set, transformasi yang sama diterapkan pada test set agar konsistensi data tetap terjaga.

```python
X_train_transformed = transformer.fit_transform(X_train)
X_test_transformed = transformer.transform(X_test)

# Mendapatkan nama kolom hasil encoding dari OneHotEncoder dan BinaryEncoder
onehot_columns = transformer.transformers_[0][1].get_feature_names_out(['Gear_Type', 'Origin', 'Options'])
binary_columns = transformer.transformers_[1][1].get_feature_names_out(['Type', 'Make', 'Region'])

# Gabungkan dengan nama kolom fitur numerik yang tidak terpengaruh oleh encoding
numerical_columns = [col for col in X.columns if col not in ['Gear_Type', 'Origin', 'Options', 'Type', 'Make', 'Region']]

# Gabungkan semua nama kolom
columns = list(onehot_columns) + list(binary_columns) + numerical_columns

# Buat DataFrame dari hasil transformasi
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=columns)

# Fungsi untuk menghitung VIF
def calc_vif(X):
    # Menambahkan konstanta untuk menghitung VIF
    X_const = add_constant(X)
    vif = pd.DataFrame()
    vif["variables"] = X_const.columns
    vif["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
    return vif

# Menghitung VIF pada X_train yang sudah ditransformasi
vif_result = calc_vif(X_train_transformed_df)

vif_result
```

## 5.2 Model Benchmark

### 5.2.1 Memilih Model Benchmark

Terdapat lima model yang dipilih untuk proses modeling, yang terdiri atas model dasar dan ensemble: Linear Regression, Ridge, Lasso, K-Nearest Neighbor, Random Forest, Gradient boost, dan XGBoost.

```python
# Define algoritma yang digunakan
lr = LinearRegression()
rr = KernelRidge(kernel='polynomial')
ls = Lasso(alpha =0.0005, random_state=randomstate)
knn = KNeighborsRegressor()
rf = RandomForestRegressor(random_state=randomstate)
gb = GradientBoostingRegressor(random_state=randomstate)
xgb = XGBRegressor(random_state=randomstate)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_rr = TransformedTargetRegressor(rr, func=np.log, inverse_func=np.exp)
log_ls = TransformedTargetRegressor(ls, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_gb = TransformedTargetRegressor(gb, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
```

```python
# Kandidat algoritma
models = [log_lr, log_rr, log_ls, log_knn, log_rf, log_gb, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []

scaling_models = [log_lr, log_rr, log_ls, log_knn]
```

#### 5.2.1.1 *Cross Validation*

```python
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=randomstate)

    if i in scaling_models:
        estimator = Pipeline([
            ('preprocessing', transformer),
            ('scaler', scaler),
            ('model', i)                 
        ])
    else:
        estimator = Pipeline([
            ('preprocessing', transformer),
            ('model', i)
        ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())

    # R²
    model_cv_r2 = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='r2', 
        error_score='raise'
        )
    score_r2.append(model_cv_r2)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())
```

```python
pd.DataFrame({
    'Model': ['Linear Regression', 'Kernel Ridge', 'Lasso', 'K-Nearest Neighbor', 'RandomForest Regressor', 'Gradient Boost', 'XGBoost Regressor'],
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae,
    'Std_MAE': nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE': nilai_std_mape,
    'Mean_R2': nilai_mean_r2, 
    'Std_R2': nilai_std_r2    
})
```

Penentuan Model terbaik dilakukan dengan mencari nilai Mean MAE, Mean RMSE, dan Mean MAPE terkecil. Semakin kecil nilai ketiga metriks evaluasi, semakin baik model melakukan prediksi.

Berdasarkan analisis dari nilai Mean RMSE, Mean MAE, dan Mean MAPE, **XGBoost Regressor** adalah model terbaik karena memiliki nilai terkecil di ketiga metrik evaluasi (RMSE, MAE, dan MAPE). Dengan demikian, **XGBoost Regressor** dapat dipertimbangkan sebagai model yang paling baik di antara yang lainnya pada tahap ini.

## 5.3 *Hyperparameter Tuning*

*Hyperparameter Tuning* dilakukan untuk menemukan kombinasi parameter terbaik yang meningkatkan model

```python
# Kedalaman pohon
max_depth = list(np.arange(3, 11))

# Learning rate
learning_rate = list(np.arange(1, 31)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 1001))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(0, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=randomstate, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=randomstate)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error', 'r2'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error',
    random_state = randomstate  
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error', 'mean_test_r2']).head()
```

```python
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

```python
xgb_tuning = random_xgb.best_estimator_

# Fitting model setelah tuning
xgb_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgb_tuning = xgb_tuning.predict(X_test)

# Hitung metrics RMSE, MAE, & MAPE setelah tuning
rmse_xgb_tuning = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuning))
mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)
mape_xgb_tuning = np.mean(np.abs((y_test - y_pred_xgb_tuning) / y_test))
r2_xgb_tuning = xgb_tuning.score(X_test, y_test)

# Simpan hasil dalam DataFrame
score_after_tuning = pd.DataFrame({
    'RMSE': [rmse_xgb_tuning],
    'MAE': [mae_xgb_tuning],
    'MAPE': [mape_xgb_tuning],
    'R2 Score': [r2_xgb_tuning]
}, index=['XGBoost Regressor'])

score_after_tuning
```
